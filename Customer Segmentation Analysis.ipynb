```python
# Check for outliers:
j=1
for i in df.columns:
    plt.subplot(6,7,j)
    sns.boxplot(df[i])
    j = j + 1
```

```python
# Handling Outliers:

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 -Q1
upper = Q3 + (1.5 * IQR)
lower = Q1 - (1.5 * IQR)
df_tra = df[(~((df<lower)|(df>upper)).any(axis =1))]
print('The Transformation of Dataset after Outlier Removal : ',df_tra.shape)
print('Original Dataset :',df.shape)
```

```python
df_tra.head(2)
```

```python
# Excluding the target column
x = df_tra.drop(['Income'], axis = 1)
y = df_tra['Income']
```

```python
# import the train_test_split module from sklearn:
from sklearn.model_selection import train_test_split
```

```python
# Let us now split the dataset into test and train
# Test Size : The proportion of the data to be included in the testing set 

x_train,x_test,y_train,y_test = train_test_split(x,y, random_state=100,
                                                test_size = 0.2)

print('x_train :',x_train.shape)
print('x_test :',x_test.shape)
print('y_train :',y_train.shape)
print('x_test :',y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Creating an instance:
ss = StandardScaler()
df_sca = ss.fit_transform(df_tra)
df_sca = pd.DataFrame(df_sca)
df_sca.head()
```

```python
from sklearn.cluster import AgglomerativeClustering, KMeans
```

```python
# Finding the Optimal Value of k
err = []
for i in range(1,10):
    km = KMeans(n_clusters = i, random_state = 100)
    km.fit(df_sca)
    err.append(km.inertia_)
```

```python
err
```

```python
# Sanity check:
plt.plot(range(1,10), err, marker = 'o')
plt.xlabel('K - Values')
plt.ylabel('Errors')
plt.title('Elbow Plot')
plt.show()
```

```python
# Finding the optimal k value using Silhoutte Score:

from sklearn.metrics import silhouette_score

for i in range(2,10):
    km2 = KMeans(n_clusters = i, random_state = 100)
    km2.fit(df_sca)
    sil_score = silhouette_score(df_sca, km2.labels_)
    print('The Silhouette Score for',i,'K - value is', sil_score)
```

```python
# Visualize the score:
from yellowbrick.cluster import SilhouetteVisualizer


for i in range(2,10):
    km2 = KMeans(n_clusters = i, random_state = 100)
    km2.fit(df_sca)
    sil_score = silhouette_score(df_sca, km2.labels_)
    
    print('The Silhouette Score for',i,'K - Value is',sil_score)
    
    sil_km2 = SilhouetteVisualizer(km2)
    sil_km2.fit(df_sca)
    plt.tight_layout()
    plt.show()
    
    
```
