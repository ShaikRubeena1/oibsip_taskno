```python
# Covarience:
cov = df.cov()
cov


# set the plot Size:
fig,ax = plt.subplots(figsize = (10,7))

# Plot a Heatmap for the Covariance Matrix
# annot : print values in each cell
# Linewidths : Specify width of the Line and specifying the Plot
# vmin : Minimum value of the Variable
# vmax : Maximum value of the Variable
# Cmap : colour Code for the Plot
# fmt = set the Decimal place of the Annot

sns.heatmap(cov,annot = True,linewidths = 0.95,
            cmap = 'copper',fmt = '.2g')

# Display the Plot:
plt.show()
```

```python
# Correlation:
corr = df.corr()
corr


# set the plot Size:
fig,ax = plt.subplots(figsize = (10,7))

# Plot a Heatmap for the Correlation Matrix
# annot : print values in each cell
# Linewidths : Specify width of the Line and specifying the Plot
# vmin : Minimum value of the Variable
# vmax : Maximum value of the Variable
# Cmap : colour Code for the Plot
# fmt = set the Decimal place of the Annot

sns.heatmap(corr,annot = True,linewidths = 0.95,
            cmap = 'gist_earth',fmt = '.2g')

# Display the Plot:
plt.show()
```

```python
df.head()
```

```python
# Set the figure Size:
plt.rcParams['figure.figsize']=[5,4]

# Plotting countplot plot for Product Category
sns.countplot(data=df,x='Product Category',color='black')

# Title of the plot
plt.title('Count Plot for Product Category')

# Displays the plot
plt.show()
```

```python
# Strip plot
sns.stripplot(x='Gender',y='Age',data=df)

# Set the title:
plt.title('Strip Plot for Gender and Age')

# display the title:
plt.show()
```

```python
# Swarm Plot:
sns.swarmplot(x='Gender',y='Age',hue='Product Category',data=df)

# Set the title
plt.title('Swarm Plot for Gender and Age')

# Displays the plot:
plt.show()
```

```python
# Barplot 
Total_Amount =df['Total Amount'].value_counts()

sns.barplot(data=df,x='Product Category',y=Total_Amount,color ='y')

# Set the title
plt.title('Bar Plot for Total Amount and Product Category')

# Display the plot
plt.show()
```

```python
# Scatter Plot:

plt.scatter(x='Total Amount',y='Age',color ='g',data=df)
plt.scatter(x='Total Amount',y='Quantity',color ='b',data=df)

plt.legend()

plt.show()
```

**Interpretation**:There is no significant correlation between **Age** and **Quantity** 

```python
# Histogram
df.plot.hist(subplots = True, layout = (3,2),figsize = (10,6),
            sharex = True,sharey = False)

# to give the secified padding from the subplot:
plt.tight_layout()

# to display the plot:
plt.show()
```

```python
# Calculating length of the Price per unit

TotalAmount = df['Total Amount']
len(TotalAmount)
```

```python
# Set the figure Size:
plt.rcParams['figure.figsize']=[8,4]

# Plotting Density plot for Total Amount
sns.distplot(TotalAmount)

# Title of the plot
plt.title('Density plot for Total Amount')

# Displays the plot
plt.show()
```

```python
PricePerUnit = df['Price per Unit']
len(PricePerUnit)
```

```python
# Set the figure Size:
plt.rcParams['figure.figsize']=[8,4]

# Plotting Density plot for Total Amount
sns.distplot(PricePerUnit,hist = False)

# Title of the plot
plt.title('Density plot for Price Per Unit')

# Displays the plot
plt.show()
```

```python
# Boxplot for Numerical Columns

df_numerical = df.describe(include = [np.number])
df_numerical.plot(kind = 'box',subplots = True,layout = (2,3))

# To give Specified padding from the subplots
plt.tight_layout()

# Display the Plot
plt.show()
```

```python
# Violinplot 
sns.violinplot(x ='Quantity',data = df, color = 'purple')

# set the title
plt.title('Violin Plot for Quantity')

# displays the plot:
plt.show()
```

```python
# Set the figure Size:
plt.rcParams['figure.figsize']=[5,4]

# Plotting Density plot for Total Amount
sns.countplot(data=df,x='Gender',color='orange')

# Title of the plot
plt.title('Count plot for Gender')

# Displays the plot
plt.show()
```
