```python
# Get the count of missing values:
missing_values = data.isnull().sum()

# check the missing values:
total = data.isnull().sum().sort_values(ascending = False)

# calculating percentage of null values:
percent = ((data.isnull().sum()/data.shape[0])*100)

# sort the values in descending order:
percent = percent.sort_values(ascending = False)

# concatenating the total missing values and percentage of missing values:
missing_data = pd.concat([total,percent],axis = 1,
                        keys = ['total','percentage'])

# add the data types:
missing_data['Type'] = data[missing_data.index].dtypes

# view the missing values:
missing_data
```

```python
# Replace all the missing values with '0'
data.description.replace(np.NaN, '0', inplace = True)
```

```python
# selecting only numerical columns:
df = data.select_dtypes(include = [np.number])
df.head(2)
```

```python
# counts all the columns:
df.columns
```

```python
# plot the boxplot for each columns:
# Subplots() : plot subplots:
# Figsize() : set the figure size:
fig,ax = plt.subplots(2,3, figsize = (15,8))

# plot the boxplot using boxplot() from seaborn library:
# z : let the variable z defines the boxplot:
# x : data from which the boxplot is to be predicted:
# orient : 'h' specifies the horizontal boxplot
# whis : proportion of the IQR, it past the low and high quartiles to extendthe plot
# ax : specifies the axes object to draw the plot
# set_xlabel() : set the x-axis label
# Fontsize() : set the font size of the x-axis
for variable, subplot in zip(df.columns, ax.flatten()):
    z = sns.boxplot(x = df[variable], orient = 'h', whis = 1.5, ax = subplot)
    z.set_xlabel(variable, fontsize = 5)
    
```

```python
# Obtain first Quartile
Q1 = df.quantile(0.25)

# Obtain third Quartile
Q3 = df.quantile(0.75)

# IQR
IQR = Q3 - Q1
print(IQR)
```

```python
# Filter out the outlier values:
data = data[~((data<(Q1-1.5*IQR))|(data>(Q3+1.5*IQR))).any(axis =1)]
```
