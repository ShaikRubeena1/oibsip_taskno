Data Cleaning is the process of fixing or removing incorrect, corrupted, duplicated or incomplete data within a dataset.

```python
# Plot heatmap to visualize the null values
plt.rcParams['figure.figsize'] = [15,5]

# Heatmap
sns.heatmap(data.isnull(), cbar = False)

# display the plot
plt.show()
```

```python
# Get the count of missing values:
missing_values = data.isnull().sum()

# check the missing values:
total = data.isnull().sum().sort_values(ascending = False)

# calculating percentage of null values:
percent = ((data.isnull().sum()/data.shape[0])*100)

# sort the values in descending order:
percent = percent.sort_values(ascending = False)

# concatenating the total missing values and percentage of missing values:
missing_data = pd.concat([total,percent],axis = 1,
                        keys = ['total','percentage'])

# add the data types:
missing_data['Type'] = data[missing_data.index].dtypes

# view the missing values:
missing_data
```

```python
# Replace all the missing values with '0'
data.description.replace(np.NaN, '0', inplace = True)
```

```python
# sanity check
data.isnull().sum().sum()
```

```python
# covariance
cov = data.cov()
cov

# set the figure size
fig,ax = plt.subplots(figsize = (10,7))

# Plot a Heatmap for the Covariance Matrix
# annot : print values in each cell
# Linewidths : Specify width of the Line and specifying the Plot
# vmin : Minimum value of the Variable
# vmax : Maximum value of the Variable
# Cmap : colour Code for the Plot
# fmt = set the Decimal place of the Annot
sns.heatmap(cov,annot = True,linewidths = 0.95,
           cmap = 'copper', fmt = '.2g')

# display the plot
plt.show()
```

```python
# correlation
corr = data.corr()
corr
# set the figure size
fig,ax = plt.subplots(figsize = (10,7))


# Plot a Heatmap for the Correlation Matrix
# annot : print values in each cell
# Linewidths : Specify width of the Line and specifying the Plot
# vmin : Minimum value of the Variable
# vmax : Maximum value of the Variable
# Cmap : colour Code for the Plot
# fmt = set the Decimal place of the Annot
sns.heatmap(corr,annot = True,linewidths = 0.95,
           cmap = 'flare', fmt = '.2g')

# display the plot
plt.show()
```

```python
# Filtering numerical columns
df = data.select_dtypes(include = [np.number])
df.head(2)
```

```python
df.columns
```

```python
# plot the boxplot for each columns:
# Subplots() : plot subplots:
# Figsize() : set the figure size:
fig,ax = plt.subplots(2,3, figsize = (15,8))

# plot the boxplot using boxplot() from seaborn library:
# z : let the variable z defines the boxplot:
# x : data from which the boxplot is to be predicted:
# orient : 'h' specifies the horizontal boxplot
# whis : proportion of the IQR, it past the low and high quartiles to extendthe plot
# ax : specifies the axes object to draw the plot
# set_xlabel() : set the x-axis label
# Fontsize() : set the font size of the x-axis

for variable, subplot in zip(df.columns, ax.flatten()):
    z = sns.boxplot(x = df[variable], orient = 'h', whis = 1.5, ax = subplot)
    z.set_xlabel(variable, fontsize = 5)
```

```python
# first Quartile
Q1 = df.quantile(0.25)
# third Quartile
Q3 = df.quantile(0.75)

# IQR
IQR = Q3 - Q1
print(IQR)
```

```python
# Filtering the outliers from the data
data = data[~((data<(Q1-1.5*IQR))|(data>(Q3+1.5*IQR))).any(axis =1)]
```

```python
data.shape
```

```python
# Removing Unnecessary data columns from the dataset
data.drop(['video_id','trending_date','title','tags','thumbnail_link','description','publish_time'],axis = 1,inplace = True)
```

```python
# Encoding
data_df = pd.get_dummies(data,columns = ['comments_disabled','ratings_disabled','video_error_or_removed'],
              drop_first= True)
data_df.head()
```
